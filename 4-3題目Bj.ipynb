```python
import pickle
with open("題目音節的分數.txt",'rb') as fp:
    score_P=pickle.load(fp)

print(score_P)
```

```python
with open("題目字長的分數.txt",'rb') as fp:
    score_L=pickle.load(fp)

print(score_L)
```

```python
from collections import OrderedDict
Bj=OrderedDict()
for i,j in score_L.items():
    Bj[i]=score_L[i]*0.7+score_P[i]*0.3
Bj=OrderedDict(sorted(Bj.items(),key=lambda t:t[1],reverse=True))

print(Bj)
```
