```
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import pickle
import numpy as np
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait

def set_up():
    # 設定一些chrome瀏覽器的變數，做了headless(不會開瀏覽器)測試，請確認chrome版本有到60
    chrome_options = Options()
    # chrome_options.add_argument("--headless")
    chrome_options.add_argument("--window-size=1920x1080")
    # 抓取selenium webdriver位置
    driver = webdriver.Chrome(options=chrome_options, executable_path= 'chromedriver')
    driver.implicitly_wait(2)  # 隱性等待，使用一次整個程式過程皆有效
    return driver

driver = set_up()

push = pd.read_excel("順序完成1113.xlsx",index_col="日期")

push.reset_index(inplace=True)

words=[]
defineday1=[]
defineday2=[]
defineday3=[]
defineday4=[]
defineday5=[]

for j in range(2):
    w=[]
    for i in range(5):
        word=push.loc[j, '單字'+str(i+1)]
        wx='https://dictionary.cambridge.org/zht/%E8%A9%9E%E5%85%B8/%E8%8B%B1%E8%AA%9E-%E6%BC%A2%E8%AA%9E-%E7%B9%81%E9%AB%94/'+word

        
        driver.get(wx)
        xpath = "//span[@class='def-body']/span[@class='trans']"
        elements = driver.find_elements_by_xpath(xpath)
        temp = []
        for i in elements:
            temp.append(i.text)
        # print(trans)
        trans = ""
        trans = ";".join(temp)
        print(trans)
        if i==0:
            defineday1.append(trans)
        elif i==1:
            defineday2.append(trans)
        elif i==2:
            defineday3.append(trans)
        elif i==3:
            defineday4.append(trans)
        elif i==4:
            defineday5.append(trans)
        # print(wx)
        w.append(trans)
    words.append(w)

with open("單字翻譯",'wb') as fp:
    pickle.dump(words,fp)

print(defineday1)
print(defineday2)
print(defineday3)
print(defineday4)
print(defineday5)



df=push
df['單字翻譯1']=defineday1
df['單字翻譯2']=defineday2
df['單字翻譯3']=defineday3
df['單字翻譯4']=defineday4
df['單字翻譯5']=defineday5


writer = ExcelWriter('推播單字www.xlsx')
df.to_excel(writer,'Sheet1',index=False)
writer.save()
```
