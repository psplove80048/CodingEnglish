```python
from pycorenlp import StanfordCoreNLP

nlp = StanfordCoreNLP('http://localhost:9000')
res = nlp.annotate("I love you. I hate him. You are nice. He is dumb",
                   properties={
                       'annotators': 'ssplit',
                       'outputFormat': 'json',
                       'timeout': 1000,
                   })

res
```

```python
for s in res["sentences"]:
    print(' '.join(t['word'] for t in s['tokens']))
```

```python
import pickle
with open("description.txt","rb") as f:
    article=pickle.load(f)
    
' '.join(article[3])
```

```python
temp = nlp.annotate(' '.join(article[3]),
                   properties={
                       'annotators': 'ssplit',
                       'outputFormat': 'json',
                       'timeout': 1000,
                   })

for s in temp["sentences"]:
    print(' '.join(t['word'] for t in s['tokens'])+'\n')
```

```python
sents=[]
for i in article:
    sent=[]
    temp = nlp.annotate(' '.join(i),
                   properties={
                       'annotators': 'ssplit',
                       'outputFormat': 'json',
                       'timeout': 1000,
                   })

    for s in temp["sentences"]:
        sent.append(' '.join(t['word'] for t in s['tokens']))
        
    sents.append(sent)
```

```python
with open('題目的斷句_CoreNLP.txt','wb') as f:
    pickle.dump(sents,f)
```
