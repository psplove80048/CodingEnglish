```python
import pickle
with open("更新Des clear word.txt",'rb') as fp:
    word_lists=pickle.load(fp)

print(word_lists[0])
```

```python
word_set=set()
for i in word_lists:
    word_set.update(i)
```

```python
word_len=[]
for i in word_set:
    word_len.append((i,len(i)))
```

```python
word_len=sorted(word_len,key=lambda t:t[1],reverse=True)
print(word_len)
```

```python
len_set=[]
for i in word_len:
    if i[1] not in len_set:
        len_set.append(i[1])
Lmax=len_set[0]
Lmid=len_set[len(len_set)//2]
Lmin=len_set[-1]
print(Lmax)
print(Lmid)
print(Lmin)
```

```python
from collections import OrderedDict
final_len=OrderedDict()
for i in word_len:
    if i[1] == Lmax:
        final_len[i[0]]=100;
    elif Lmax>i[1]>Lmid:
        final_len[i[0]]=round(50+(i[1]-Lmid)*(100-50)/(Lmax-Lmid),3)
    elif i[1] == Lmid:
        final_len[i[0]]=50;
    elif Lmid>i[1]>Lmin:
        final_len[i[0]]=round(0+(i[1]-Lmin)*(50-0)/(Lmid-Lmin),3)
    else :
        final_len[i[0]]=0
```

```python
final_len=OrderedDict(sorted(final_len.items(),key=lambda t:t[1],reverse=True))
print(final_len)
```

```python
with open('題目字長的分數.txt','wb') as f:
    pickle.dump(final_len,f)
f.close()
```
