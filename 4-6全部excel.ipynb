```python
# 變數
# 上個題目單字數量之mod5餘數
extra = 0
today = date.today()

for i in X1:
    # uva題目編號
    index_of_uva = i[1]
    # 自用索引值
    index = i[0]
    # 一道題目中的所有單字
    one_question = word[index]
    one_question_sent = sentence[index]
    one_quiz = new_all_quizs[index]
    one_answer = new_all_answers[index]
    # 一道題目中的所有單字數量
    number_of_questions = len(one_question)
    
    # print(col_vocab1)
#     print(one_question)
#     print(number_of_questions%5)

    for i in range(len(one_question)):
        x = i + extra
        if x%5==0:
            col_vocab1.append(one_question[i])
            sentences_one.append(one_question_sent[i])
            col_vocab_quiz1.append(one_quiz[i])
            quiz1_answer.append(one_answer[i])
        elif x%5==1:
            col_vocab2.append(one_question[i])
            sentences_two.append(one_question_sent[i])
            col_vocab_quiz2.append(one_quiz[i])
            quiz2_answer.append(one_answer[i])
        elif x%5==2:
            col_vocab3.append(one_question[i])
            sentences_three.append(one_question_sent[i])
            col_vocab_quiz3.append(one_quiz[i])
            quiz3_answer.append(one_answer[i])
        elif x%5==3:
            col_vocab4.append(one_question[i])
            sentences_four.append(one_question_sent[i])
            col_vocab_quiz4.append(one_quiz[i])
            quiz4_answer.append(one_answer[i])
        elif x%5==4:
            col_vocab5.append(one_question[i])
            sentences_five.append(one_question_sent[i])
            col_vocab_quiz5.append(one_quiz[i])
            quiz5_answer.append(one_answer[i])
            col_question_num.append(index_of_uva)

    # 題目單字數量之mod5餘數
    extra = number_of_questions%5 + extra
    # print(extra)

cal = extra%5
if cal==1:
    del col_vocab1[len(col_vocab1)-1]
    del sentences_one[len(sentences_one)-1]
    del col_vocab_quiz1[len(col_vocab_quiz1)-1]
    del quiz1_answer[len(quiz1_answer)-1]    
elif cal==2:
    del col_vocab1[len(col_vocab1)-1]
    del col_vocab2[len(col_vocab2)-1]
    del sentences_one[len(sentences_one)-1]
    del sentences_two[len(sentences_two)-1]
    del col_vocab_quiz1[len(col_vocab_quiz1)-1]
    del col_vocab_quiz2[len(col_vocab_quiz2)-1]
    del quiz1_answer[len(quiz1_answer)-1]
    del quiz2_answer[len(quiz2_answer)-1]
elif cal==3:
    del col_vocab1[len(col_vocab1)-1]
    del col_vocab2[len(col_vocab2)-1]
    del col_vocab3[len(col_vocab3)-1]
    del sentences_one[len(sentences_one)-1]
    del sentences_two[len(sentences_two)-1]
    del sentences_three[len(sentences_three)-1]
    del col_vocab_quiz1[len(col_vocab_quiz1)-1]
    del col_vocab_quiz2[len(col_vocab_quiz2)-1]
    del col_vocab_quiz3[len(col_vocab_quiz3)-1]
    del quiz1_answer[len(quiz1_answer)-1]
    del quiz2_answer[len(quiz2_answer)-1]
    del quiz3_answer[len(quiz3_answer)-1]
elif cal==4:
    del col_vocab1[len(col_vocab1)-1]
    del col_vocab2[len(col_vocab2)-1]
    del col_vocab3[len(col_vocab3)-1]
    del col_vocab4[len(col_vocab4)-1]
    del sentences_one[len(sentences_one)-1]
    del sentences_two[len(sentences_two)-1]
    del sentences_three[len(sentences_three)-1]
    del sentences_four[len(sentences_four)-1]
    del col_vocab_quiz1[len(col_vocab_quiz1)-1]
    del col_vocab_quiz2[len(col_vocab_quiz2)-1]
    del col_vocab_quiz3[len(col_vocab_quiz3)-1]
    del col_vocab_quiz4[len(col_vocab_quiz4)-1]
    del quiz1_answer[len(quiz1_answer)-1]
    del quiz2_answer[len(quiz2_answer)-1]
    del quiz3_answer[len(quiz3_answer)-1]
    del quiz4_answer[len(quiz4_answer)-1]

print("mission complete~")
print(len(col_vocab1))
print(len(col_vocab2))
print(len(col_vocab3))
print(len(col_vocab4))
print(len(sentences_one))
print(len(sentences_two))
print(len(sentences_three))
print(len(sentences_four))
print(len(col_vocab_quiz1))
print(len(col_vocab_quiz2))
print(len(col_vocab_quiz3))
print(len(col_vocab_quiz4))
print(len(quiz1_answer))
print(len(quiz2_answer))
print(len(quiz3_answer))
print(len(quiz4_answer))
```

```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
data={'日期':days,'單字1':col_vocab1,'單字2':col_vocab2,'單字3':col_vocab3,'單字4':col_vocab4,'單字5':col_vocab5,'題目編號':col_question_num,'句子1':sentences_one,'句子2':sentences_two,'句子3':sentences_three,'句子4':sentences_four,'句子5':sentences_five,'題目1':col_vocab_quiz1,'題目2':col_vocab_quiz2,'題目3':col_vocab_quiz3,'題目4':col_vocab_quiz4,'題目5':col_vocab_quiz5,'答案1':quiz1_answer,'答案2':quiz2_answer,'答案3':quiz3_answer,'答案4':quiz4_answer,'答案5':quiz5_answer}
df = pd.DataFrame(data,columns=['日期','單字1','單字2','單字3','單字4','單字5','題目編號','句子1','句子2','句子3','句子4','句子5','答案1','答案2','答案3','答案4','答案5',])
```

```python
import random
for i in df.index.get_level_values(0):   
    a=new_all_Fillinblanks[i]
    for b in range(len(a)):
        df.at[i,'填空題'+str(b+1)]=a[b]
        
for i in df.index.get_level_values(0):
    d= new_all_sents[i] 
    for b in range(len(d)):
        df.at[i,'每日句子'+str(b+1)]=d[b]
for i in df.index.get_level_values(0):
    items = [1, 2, 3, 4, 5]
    random.shuffle(items)
    randnum=''.join(str(v) for v in items)
    df.at[i,'亂數順序']=randnum
for i in df.index.get_level_values(0):
    volcount= new_all_answer_Fillinblanks[i] 
    for b in range(len(volcount)):
        df.at[i,'單字數每日句子'+str(b+1)]=str(len(volcount[b]))
for i in df.index.get_level_values(0):
    for b in range(5):
        if df.at[i,'單字數每日句子'+str(b+1)] == ('NaN'):
            df.at[i,'單字數每日句子'+str(b+1)]=0
for i in df.index.get_level_values(0):
    items = [1, 2, 3, 4, 5]
    random.shuffle(items)
    randnum=''.join(str(v) for v in items)
    df.at[i,'亂數順序1']=randnum
```
