前處理 stopword去標點和數字

```
import string
import nltk
from nltk.corpus import stopwords
import re
wordsfiltered=[]


#小寫化->word_list
word_list=[]
for words in word_lists:
    word_list.append([i.lower() for i in words])

#
stop_words=stopwords.words('english')

r='[’‘•!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]+'

for i in word_list:
    temp=[x for x in i if x not in stop_words and not re.search('[0-9]+',x) and not re.search(r,x) and '\\' not in x and len(x)>4]
    wordsfiltered.append(temp)

print('success')
```

用wordnet踢出不是單字的文字

```
from nltk.corpus import wordnet

word_to_test="vice"
if wordnet.synsets(word_to_test):
  print("English Word")
```

詞幹還原 前處理結束

```
import nltk

lmtzr=nltk.WordNetLemmatizer()
array_lmtzr=[]

for count in range(len(wordsfiltered)):
    #temp=[lmtzr.lemmatize(i) for i in wordsfiltered[count]]
    temp=[]
    for word_list, tag in nltk.pos_tag(wordsfiltered[count]):
        if tag.startswith('NN'):
            temp.append(lmtzr.lemmatize(word_list,pos='n'))
        elif tag.startswith('VB'):
            temp.append(lmtzr.lemmatize(word_list,pos='v'))
        elif tag.startswith('JJ'):
            temp.append(lmtzr.lemmatize(word_list,pos='a'))
        elif tag.startswith('R'):
            temp.append(lmtzr.lemmatize(word_list,pos='r'))
        else:
            temp.append(word_list)
    array_lmtzr.append(temp)
    
print('success')
```

TF-IDF 分析

```
from collections import OrderedDict
order=OrderedDict()
for i in range(len(clear_text)):
    tc=nltk.TextCollection(clear_text[i])
    for word_list in clear_text[i]:
        order[word_list]=tc.tf_idf(word_list,clear_text[i])
order=OrderedDict(sorted(order.items(),key=lambda t:t[1],reverse=True))
print(order)
```
