```python
from pyphonetics import Metaphone
metaphone = Metaphone()
metaphone.phonetics('banana')
```

```python
import pickle
with open("更新Des clear word.txt",'rb') as fp:
    word_lists=pickle.load(fp)

print(word_lists[0])
```

```python
word_set=set()
for i in word_lists:
    word_set.update(i)
```

# **計算音節長度**

```python
word_PHlen=[]
for i in word_set:
    word_PHlen.append((i,len(metaphone.phonetics(i))))
```

```python
word_PHlen=sorted(word_PHlen,key=lambda t:t[1],reverse=True)
print(word_PHlen)
```

# 計算音節分數 四捨五入

```python
word_PR=[]
for i in word_PHlen:
    word_PR.append((i[0],round(len(i[0])/i[1],3)))
word_PR=sorted(word_PR,key=lambda t:t[1],reverse=True)
print(word_PR)
```

```python
score_set=[]
for i in word_PR:
    if i[1] not in score_set:
        score_set.append(i[1])
print(score_set)
PRmax=score_set[0]
PRmin=score_set[-1]
```

```python
from collections import OrderedDict
final_PR=OrderedDict()
for i in word_PR:
    if i[1] == PRmax:
        final_PR[i[0]]=100;
    elif i[1] == PRmin:
        final_PR[i[0]]=0;
    else:
        final_PR[i[0]]=round(((100-0)/(PRmax-PRmin))*(i[1]-PRmin),3)
```

```python
final_PR=OrderedDict(sorted(final_PR.items(),key=lambda t:t[1],reverse=True))
print(final_PR)
```

```python
with open('題目音節的分數.txt','wb') as f:
    pickle.dump(final_PR,f)
f.close()
```
